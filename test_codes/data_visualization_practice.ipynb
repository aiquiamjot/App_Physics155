```python
import matplotlib.pyplot as plt 
import numpy as np

days = ['Sun','Mon','Tue','Wed','Thu','Fri','Sat']
win_loss = [[23,12],[16,24],[17,11],[21,21],[25,21],[21,20],[25,20]]
WL_array = np.array(win_loss)
wins = WL_array[:,0]
loss = WL_array[:,1]

x_axis = np.arange(len(days))
plt.bar(x_axis - 0.2, wins, 0.4, label = 'Wins', color = '#4CAF50')
plt.bar(x_axis + 0.2, loss, 0.4, label = 'Loss', color = 'hotpink')

plt.xticks(x_axis, days)
plt.xlabel('Days of the Week')
plt.ylabel('Number of Games')
plt.title('Dota Games Won and Lost Every Day of the Week in the Past 3 Months')
plt.legend()
plt.show()

```
